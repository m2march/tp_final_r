```r
h = function(a) {
    xs = c(-1, 1, 4, 5, 7, 10)
    ps = c(2/24, 6/24, 4/24, 1/24, 7/24, 4/24)
    return(sum(((xs - a) ** 2) * ps))
}
h.median = sum(c(-1, 1, 4, 5, 7, 10) * c(2/24, 6/24, 4/24, 1/24, 7/24, 4/24))
```

```r
xs = seq(-3, 11, length=100)
ys = sapply(xs, h)
plot(xs, ys, type='l', col='blue')
par(new=T)
lines(c(h.median, h.median), c(0, max(ys)), col='black')
```
