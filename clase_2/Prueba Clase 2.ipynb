```r
normfit <- function(col, name) {
    media <- mean(col)
    desvio <- sd(col)
    h <- hist(col, plot=F)

    maxh <- max(h$density) + 0.0005
    
    xs <- seq(min(col), max(col), length=100)
    ys <- dnorm(xs, media, desvio)
    
    hist(col, freq=F, ylim=c(0, maxh), main=name, xlab='SAT')
    lines(xs, ys, col='blue')    
}

par(mfrow=c(1,2))
normfit(VerbalSAT, 'VerbalSAT')
normfit(MathSAT, 'MathSAT')
```
